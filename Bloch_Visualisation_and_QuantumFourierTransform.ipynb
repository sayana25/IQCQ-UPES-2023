The quantum Fourier transform (QFT) is the quantum implementation of the discrete Fourier transform over the amplitudes of a wavefunction. It is a significant step in many quantum algorithms, most notably Shor's factoring algorithm and quantum phase estimation. The quantum Fourier transform acts on a quantum state $|X\rangle = \sum\limits_{j = 0}^{N-1}x_{j}|j\rangle$ mapping it to the quantum state $|Y\rangle = \sum\limits_{j = 0}^{N-1}y_{k}|k\rangle$ according to the formula,

$y_{k} =\frac{1}{\sqrt{N}}\sum\limits_{j = 0}^{N-1} x_{j}e^{2\pi i \frac{jk}{N}}$. This transformation could infact be represented by an unitary matrix,
$U_{QFT} = \frac{1}{\sqrt{N}} \sum\limits_{j = 0}^{N-1}\sum\limits_{k = 0}^{N-1} e^{2\pi i \frac{jk}{N}} |k\rangle\langle j|$.

For implementing the QFT circuit, two gates are required, the Hadamard and the two-qubit controlled rotation $CROT_{k}$ gate. In QISKIT this is represented by the $CP(\theta)$.

```python
import numpy as np
from numpy import pi
# importing Qiskit
from qiskit import QuantumCircuit, transpile, Aer, IBMQ
from qiskit.tools.monitor import job_monitor
from qiskit.visualization import plot_histogram, plot_bloch_multivector
```

**3-qubit QFT**

```python
qc = QuantumCircuit(3)
qc.h(2)
qc.cp(pi/2, 1, 2) # CROT from qubit 1 to qubit 2
qc.cp(pi/4, 0, 2) # CROT from qubit 2 to qubit 0
qc.h(1)
qc.cp(pi/2, 0, 1) # CROT from qubit 0 to qubit 1
qc.h(0)
qc.swap(0,2)
qc.draw('mpl')
```

**General number of qubits in the QFT**, shown for 4 qubits in the example

```python
def qft_rotations(circuit, n):
    """Performs qft on the first n qubits in circuit (without swaps)"""
    if n == 0:
        return circuit
    n -= 1
    circuit.h(n)
    for qubit in range(n):
        circuit.cp(pi/2**(n-qubit), qubit, n)
    # At the end of our function, we call the same function again on
    # the next qubits (we reduced n by one earlier in the function)
    qft_rotations(circuit, n)

# Let's see how it looks:
qc = QuantumCircuit(4)
qft_rotations(qc,4)
qc.draw('mpl')
```

**General QFT circuit**

```python
def swap_registers(circuit, n):
    for qubit in range(n//2):
        circuit.swap(qubit, n-qubit-1)
    return circuit

def qft(circuit, n):
    """QFT on the first n qubits in circuit"""
    qft_rotations(circuit, n)
    swap_registers(circuit, n)
    return circuit

# Let's see how it looks:
qc = QuantumCircuit(4)
qft(qc,4)
qc.draw('mpl')
```
